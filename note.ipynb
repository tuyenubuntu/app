Flask app
1. API là gì?
API (Application Programming Interface) là một giao diện cho phép các ứng dụng phần mềm giao tiếp và trao đổi dữ liệu với nhau. API cung cấp các phương thức hoặc quy trình mà một ứng dụng có thể sử dụng để yêu cầu dữ liệu từ ứng dụng khác hoặc gửi dữ liệu tới ứng dụng khác mà không cần biết chi tiết cách hoạt động bên trong của nó.

Ví dụ, khi bạn sử dụng ứng dụng di động để xem thời tiết, ứng dụng này có thể gửi yêu cầu (request) tới một API thời tiết, nhận dữ liệu từ đó và hiển thị kết quả cho bạn.

Các loại API phổ biến:

Web API: API được xây dựng trên nền tảng web, cho phép giao tiếp qua các giao thức HTTP, thường trả về dữ liệu ở định dạng JSON hoặc XML.
RESTful API: Là một phong cách thiết kế cho các API web, sử dụng các phương thức HTTP (như GET, POST, PUT, DELETE) để thực hiện các thao tác với tài nguyên.
2. JSON là gì?
JSON (JavaScript Object Notation) là một định dạng dữ liệu phổ biến, dễ đọc và dễ viết, dùng để trao đổi dữ liệu giữa các ứng dụng hoặc giữa máy khách và máy chủ. JSON thường được sử dụng trong các API vì nó nhẹ nhàng, dễ dàng phân tích cú pháp và thao tác bằng nhiều ngôn ngữ lập trình.

Ví dụ về một đối tượng JSON:

json
Copy code
{
  "name": "John Doe",
  "age": 30,
  "city": "New York"
}
Định dạng JSON bao gồm:

Các cặp khóa - giá trị (key-value pairs)
Giá trị có thể là số, chuỗi, mảng, đối tượng khác hoặc giá trị boolean (true/false).
3. GET là gì?
GET là một phương thức HTTP (Hypertext Transfer Protocol) được sử dụng để yêu cầu dữ liệu từ một server. Khi một yêu cầu GET được gửi đến server, server sẽ trả về dữ liệu (nội dung của trang, JSON, HTML, hình ảnh, v.v.) mà không thay đổi bất kỳ dữ liệu nào trên server.

Ví dụ:

GET /users: Yêu cầu danh sách tất cả người dùng từ server.
GET /users/1: Yêu cầu thông tin người dùng có ID là 1.
4. Các HTTP Methods phổ biến
Trong giao tiếp giữa client và server, các phương thức HTTP thường được sử dụng để tương tác với tài nguyên (ví dụ như trang web, dữ liệu, hoặc các dịch vụ). Dưới đây là các phương thức HTTP cơ bản:

GET: Dùng để yêu cầu dữ liệu từ server. Phương thức này chỉ lấy dữ liệu mà không làm thay đổi dữ liệu trên server.

Ví dụ: GET /users để lấy danh sách người dùng.
POST: Dùng để gửi dữ liệu từ client lên server, thường dùng khi tạo mới một tài nguyên.

Ví dụ: POST /users để tạo một người dùng mới.
PUT: Dùng để cập nhật dữ liệu của một tài nguyên trên server. Nếu tài nguyên chưa tồn tại, phương thức này sẽ tạo mới.

Ví dụ: PUT /users/1 để cập nhật thông tin của người dùng có ID = 1.
DELETE: Dùng để xóa một tài nguyên khỏi server.

Ví dụ: DELETE /users/1 để xóa người dùng có ID = 1.
PATCH: Dùng để cập nhật một phần dữ liệu của tài nguyên. Khác với PUT, PATCH chỉ cập nhật một phần tài nguyên chứ không phải toàn bộ.

Ví dụ: PATCH /users/1 để cập nhật một số thông tin của người dùng có ID = 1.
OPTIONS: Dùng để yêu cầu thông tin về các phương thức HTTP mà server hỗ trợ cho một tài nguyên.

Ví dụ: OPTIONS /users để kiểm tra các phương thức HTTP mà server hỗ trợ cho tài nguyên người dùng.
HEAD: Giống như GET, nhưng không trả về nội dung của tài nguyên, chỉ trả về các tiêu đề HTTP (headers). Thường dùng để kiểm tra trạng thái của tài nguyên mà không cần tải toàn bộ dữ liệu.

Ví dụ: HEAD /users/1 để kiểm tra xem người dùng có tồn tại mà không tải dữ liệu của người dùng đó.
Tổng kết
API giúp các ứng dụng giao tiếp và trao đổi dữ liệu.
JSON là định dạng phổ biến để trao đổi dữ liệu giữa client và server.
GET là phương thức HTTP dùng để yêu cầu dữ liệu từ server mà không làm thay đổi dữ liệu.
Các phương thức HTTP khác như POST, PUT, DELETE, PATCH cũng đóng vai trò quan trọng trong việc thao tác với tài nguyên trên server.

Trong thư viện Flask, bạn sẽ thấy các thành phần được nhập khẩu như Flask, render_template, request, jsonify, v.v. Mỗi thành phần này có một mục đích sử dụng riêng trong việc phát triển ứng dụng web. Dưới đây là giải thích về khi nào và tại sao bạn cần dùng từng thành phần này:

1. Flask
Khi nào sử dụng:

Bạn cần sử dụng Flask khi tạo một ứng dụng Flask. Đây là lớp chính để khởi tạo ứng dụng của bạn.
Ví dụ sử dụng:

python
Copy code
from flask import Flask
app = Flask(__name__)
Giải thích: Flask là lớp chính trong thư viện Flask và được sử dụng để khởi tạo ứng dụng web. Bạn cần tạo một instance của lớp Flask để xây dựng ứng dụng.

2. render_template
Khi nào sử dụng:

Bạn sử dụng render_template khi muốn Flask hiển thị các file HTML, thường là khi ứng dụng của bạn có giao diện người dùng (UI) và bạn cần trả về các trang web động (rendering HTML templates).
Ví dụ sử dụng:

python
Copy code
from flask import render_template

@app.route('/')
def home():
    return render_template('index.html', title='Trang Chủ')
Giải thích:

render_template giúp Flask tìm và trả về các file HTML trong thư mục templates. Bạn có thể truyền các biến (như title trong ví dụ trên) vào HTML để render nội dung động. Đây là cách Flask xử lý giao diện người dùng.
Khi cần dùng: Khi bạn muốn render các trang HTML và truyền dữ liệu từ backend (Server) vào giao diện web.

3. request
Khi nào sử dụng:

Bạn cần sử dụng request khi muốn truy cập dữ liệu từ yêu cầu của client, ví dụ như khi lấy tham số từ URL, dữ liệu từ form, hoặc body của yêu cầu HTTP.
Ví dụ sử dụng:

python
Copy code
from flask import request

@app.route('/greet', methods=['POST'])
def greet():
    name = request.form['name']  # Lấy dữ liệu từ form gửi lên
    return f"Xin chào, {name}!"
Giải thích: request cho phép bạn truy cập dữ liệu gửi từ client tới server, bao gồm:

request.form: Dữ liệu từ các form POST.
request.args: Dữ liệu từ URL (query parameters) trong các yêu cầu GET.
request.json: Dữ liệu JSON trong body của yêu cầu.
Khi cần dùng: Khi bạn muốn truy cập dữ liệu từ yêu cầu HTTP, ví dụ như lấy thông tin từ form, URL, hoặc JSON gửi lên server.

4. jsonify
Khi nào sử dụng:

Bạn sử dụng jsonify khi muốn trả về dữ liệu dưới dạng JSON từ server. Đây là cách phổ biến khi xây dựng các API RESTful, nơi bạn cần trả về dữ liệu có cấu trúc JSON.
Ví dụ sử dụng:

python
Copy code
from flask import jsonify

@app.route('/api/data', methods=['GET'])
def get_data():
    data = {"message": "Hello, World!", "status": "success"}
    return jsonify(data)
Giải thích: jsonify giúp bạn dễ dàng chuyển đổi các đối tượng Python (như dict, list) thành phản hồi JSON hợp lệ với đúng header HTTP. Điều này rất hữu ích khi bạn xây dựng API RESTful.

Khi cần dùng: Khi bạn muốn trả về dữ liệu dưới dạng JSON từ server, thường dùng trong các API.

5. url_for
Khi nào sử dụng:

Bạn sử dụng url_for khi cần tạo URL cho các route trong ứng dụng. Điều này rất hữu ích khi bạn muốn tạo các liên kết động trong HTML hoặc redirect trong Flask.
Ví dụ sử dụng:

python
Copy code
from flask import url_for, redirect

@app.route('/')
def home():
    return redirect(url_for('about'))  # Điều hướng tới route '/about'

@app.route('/about')
def about():
    return "Trang Giới Thiệu"
Giải thích: url_for tự động sinh ra URL cho các route trong Flask, giúp bạn tránh việc phải viết các URL tĩnh. Nếu bạn thay đổi tên route sau này, url_for sẽ tự động cập nhật liên kết mà không phải chỉnh sửa nhiều nơi trong mã nguồn.

Khi cần dùng: Khi bạn cần tạo URL cho các route của ứng dụng, hoặc khi bạn cần redirect đến một route khác.

6. session
Khi nào sử dụng:

Bạn sử dụng session để lưu trữ dữ liệu giữa các lần yêu cầu của người dùng, giống như cookies. Đây là cách để lưu trữ thông tin người dùng (ví dụ, đăng nhập, giỏ hàng) trong suốt quá trình sử dụng ứng dụng.
Ví dụ sử dụng:

python
Copy code
from flask import session

@app.route('/login', methods=['POST'])
def login():
    session['username'] = request.form['username']
    return "Đăng nhập thành công!"

@app.route('/profile')
def profile():
    username = session.get('username')
    return f"Chào {username}!"
Giải thích: session lưu trữ thông tin người dùng trong suốt phiên làm việc. Dữ liệu được mã hóa và lưu trong cookies trên client, giúp duy trì trạng thái giữa các yêu cầu HTTP.

Khi cần dùng: Khi bạn cần lưu trữ thông tin trong phiên người dùng, ví dụ như khi quản lý đăng nhập, giỏ hàng, hoặc các thông tin trạng thái khác.

Tóm tắt:
Flask: Sử dụng khi khởi tạo ứng dụng Flask.
render_template: Dùng để render các trang HTML động.
request: Dùng để truy xuất dữ liệu từ yêu cầu client (form, URL, JSON, v.v.).
jsonify: Dùng khi bạn cần trả về dữ liệu dưới dạng JSON từ server, thường trong API.
url_for: Dùng để tạo URL cho các route hoặc redirect.
session: Dùng để lưu trữ và truy xuất dữ liệu giữa các lần yêu cầu, thường dùng trong quản lý trạng thái người dùng.
